```python
#production price management system(array)
import numpy as np
arr =np.array([1000, 1200, 1500, 2000, 2300, 2500, 2600, 3000, 500, 190, 300, 310, 750, 800, 950])
print(arr)
print(np.sum(arr))
print(np.mean(arr))
print(np.max(arr))
print(np.min(arr))
```

```python
# Count products
print(np.array(arr))
above1000 = 0
below500 = 0

for price in arr:
    if price > 1000:
        above1000 += 1
    if price < 500:
        below500 += 1

print("Products > 1000:", above1000)
print("Products < 500:", below500)
```

```python
print(np.array(arr))
search = int(input("Enter price to search: "))

if search in arr:
    print("Price Found")
else:
    print("Price Not Found")
```

```python
#ascending
print(np.array(arr))
print(np.sort(arr))
```

```python
#replace product price new value
print(np.array(arr))
old = int(input("Enter old price: "))
new = int(input("Enter new price: "))
```

```python
#final price
print(np.array(arr))
arr[arr == old] = new
print(arr)
```
